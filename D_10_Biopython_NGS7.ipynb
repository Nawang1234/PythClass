```python
from Bio.Seq import Seq
dna_seq=Seq("ACTGCAGATACGAT")
#Transcribe DNA to RNA
rna_seq=dna_seq.transcribe()
#translate
prt_seq=rna_seq.translate()
rna_seq
#Complement
cDNA=dna_seq.complement()
cDNA
# reverse
rev_compl=dna_seq.reverse_complement()
rev_compl
print("DNA:",dna_seq)
print("RNA:",rna_seq)
print("PRT:",prt_seq)
print("CDNA:",cDNA)
```

```python
from Bio import SeqIO
fasta_file="/content/demo.fasta"
# parse
records=SeqIO.parse(fasta_file,"fasta")
# Iterate -id,desc,seq
for record in records:
  print("ID:",record.id)
  print("Desc:",record.description)
  print("Sequence:",record.seq)
```

```python
from Bio import pairwise2
seq1=Seq("ATGCAGTAGACAGT")
seq2=Seq("ATGACAGCAGTAGT")

#alignment
alignmets=pairwise2.align.globalxx(seq1,seq2,one_alignment_only=True)

for base in alignmets:
  print("Alignmnet:",base.seqA)
  print("Alignmnet:",base.seqB)
  print(f'Score:{base.score}')
```
