```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import csv
import time

# Use WebDriver Manager to automatically set up ChromeDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open the Wikipedia page for Amazon
url = "https://en.wikipedia.org/wiki/Amazon_(company)"
driver.get(url)

# Wait for the page to load
time.sleep(5)

# Locate the table containing the country and language information
try:
    table = driver.find_element(By.CLASS_NAME, "wikitable")  # Adjust the class name if necessary
    rows = table.find_elements(By.TAG_NAME, "tr")
except Exception as e:
    print(f"Error locating table: {e}")
    rows = []

# Extract "Country" and "Language(s)" data
country_language_data = []
for row in rows[1:]:  # Skip the header row
    try:
        columns = row.find_elements(By.TAG_NAME, "td")
        if len(columns) >= 2:
            country = columns[0].text.strip()
            languages = columns[1].text.strip()
            country_language_data.append((country, languages))
    except Exception as e:
        print(f"Error extracting data from row: {e}")
        continue

# Print the extracted data
for country, languages in country_language_data:
    print(f"Country: {country}, Languages: {languages}")

# Save the data to a CSV file
with open("amazon_country_languages.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Country", "Language(s)"])
    writer.writerows(country_language_data)

print("Data saved to 'amazon_country_languages.csv'")

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager
import csv
import time

# Set up WebDriver using WebDriver Manager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open the Wikipedia page for Amazon
url = "https://en.wikipedia.org/wiki/Amazon_(company)"
driver.get(url)

# Wait for the page to load
time.sleep(5)

# Locate the table containing "Year" and "Number of Employees"
try:
    # Find the specific table containing employee data using XPath
    table = driver.find_element(By.XPATH, "//table[contains(@class, 'wikitable')]")
    
    # Get all rows of the table
    rows = table.find_elements(By.TAG_NAME, "tr")
    
    # Extract all data (including headers)
    table_data = []
    for row in rows:
        columns = row.find_elements(By.TAG_NAME, "td")
        header_columns = row.find_elements(By.TAG_NAME, "th")
        
        # If header cells exist, it's the header row
        if header_columns:
            headers = [header.text.strip() for header in header_columns]
            table_data.append(headers)
        elif columns:
            row_data = [column.text.strip() for column in columns]
            table_data.append(row_data)

except NoSuchElementException as e:
    print(f"Error locating table or extracting data: {e}")
    table_data = []

# Print extracted table data
for row in table_data:
    print(row)

# Save the data to a CSV file
with open("amazon_employee_data.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerows(table_data)

print("Data saved to 'amazon_employee_data.csv'")

# Close the browser
driver.quit()
```

```python
# Import required libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager
import csv
import time

# Use WebDriver Manager to set up the WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open the target website
url = "https://en.wikipedia.org/wiki/Amazon_(company)"
driver.get(url)

# Wait for the page to load
time.sleep(5)  # Use WebDriverWait in production for better handling of dynamic content

# Extract "Country" and "Languages" from the infobox
try:
    # Find the infobox table
    infobox = driver.find_element(By.CLASS_NAME, "wikitable")
    
    # Extract "Country" data using accurate XPath
    country_element = infobox.find_element(By.CLASS_NAME, "headersort")
    country = country_element.text.strip() if country_element else "Not Available"
    
    # Extract "Languages" data using accurate XPath
    language_element = infobox.find_element(By.CLASS_NAME, "headersort")
    languages = language_element.text.strip() if language_element else "Not Available"
    
except NoSuchElementException as e:
    print(f"Error extracting data: {e}")
    country = "Not Available"
    languages = "Not Available"

# Print the data
print(f"Country: {country}")
print(f"Languages: {languages}")

# Save to CSV
with open("amazon_country_languages.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Country", "Languages"])
    writer.writerow([country, languages])

print("Data saved to amazon_country_languages.csv")

# Close the browser
driver.quit()
```
