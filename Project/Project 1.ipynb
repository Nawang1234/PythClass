```python
# Using IQR to detect outliers in sales

Q1 = hotel_data['CREDIT'].quantile(0.25)
Q3 = hotel_data['CREDIT'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
outliers = hotel_data[(hotel_data['CREDIT'] < lower_bound) | (hotel_data['CREDIT'] > upper_bound)]

# Plotting
plt.boxplot(hotel_data['CREDIT'])
plt.title('Outlier Detection in Sales')
plt.ylabel('Sales')
plt.show()
```
