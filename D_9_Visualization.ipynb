```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Scatter plot with customizations
plt.scatter(x, y, color="red", edgecolor="black", s=200, marker='*', alpha=0.6)

# Line plot with customizations
plt.plot(x, y, color="blue", linestyle='-.', linewidth=3, marker='o', markersize=10, alpha=0.7)

# Adding title and labels
plt.title("Scatter and Line Plot Example")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the plot
plt.show()
```

```python
gene=['GeneA','GeneB',"GeneC",'GeneD']
gene
exp_val=np.random.rand(len(gene))*100
exp_val
# generate lineplot
plt.plot(gene,exp_val,marker="o",linestyle="-")
plt.title("gene-Expression")
plt.xlabel("Genes")
plt.ylabel("Gene Expression Profile")
plt.grid(True)                  # Add grid
plt.show()                      # Display plot

```

```python
# BAR
x=[6,7,9]
y=[10,12,14]
x1=[5,8,11]
y1=[13,15,18]

plt.bar(x,y,color="red")
plt.bar(x1,y1)
plt.title("BAR Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
seq="ATGACGACAGTAGACAGTAGAC"
# calcuate freq
nucl_count={'A':seq.count('A'),
            'G':seq.count('G'),
            'T':seq.count('T'),
            'C':seq.count('C')
            }
nucl_count
plt.bar(nucl_count.keys(),nucl_count.values(),color=['blue','green','red','purple'])
plt.show()
```

```python
# Pie
# Data to plot
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)  # explode 1st slice

# Create a pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Display the plot
plt.show()
```
