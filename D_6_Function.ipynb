```python
#arbitrary
def sum_numbers(*args):
  total=0
  for num in args:
    total+=num
  return total
```

```python
sum_numbers(1,2,3,4,5)
sum_numbers(10,20,30,40,50,60,70,80,90)
```

```python
def greet_people(**kwargs):
  for key,value in kwargs.items():
    print(f"Hello {key}! You are {value} yeats old.")
```

```python
greet_people(seeta=30)
greet_people(seeta=30,Geeta=20,teena=10,Meena=40)
```

```python
#5!=5*4*3*2*1
def factorial (x):
  if x==0 or x==1:
    return 1
  else:
    return x*factorial(x-1)
```

```python
factorial(5)
```
