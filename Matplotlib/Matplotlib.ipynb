```python
np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b',c='c', s='d',data = data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 3), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the Axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the Axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x_label')  # Add an x-label to the Axes.
ax.set_ylabel('y_label')  # Add a y-label to the Axes.
ax.set_title("hARRY'S Plot")  # Add a title to the Axes.
ax.legend()  # Add a legend.
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) Axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snsFv
data = pd.read_csv("iris.csv")
data.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['petal_length'],label = 'Petal Length')
plt.plot(data['petal_width'],label = 'Petal Width')
plt.title("Petal Length and Width")
plt.xlabel("Index")
plt.ylabel("Length/Width")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data['petal_length'],data['petal_width'],color = 'green')
plt.title("Scatter plot of petal length vs width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(data['petal_length'],bins=10,color='purple',alpha = 0.5)
plt.title("Distrubution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
species_count = data['species'].value_counts()
plt.figure(figsize=(10,5))
species_count.plot(kind='bar',color=['blue','orange','green'])
plt.title("Species Counts")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
data.boxplot(column='petal_length',by='species')
plt.title("Boxplot of petal length by species")
plt.suptitle('')
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.show()
```

```python
plt.figure(figsize=(10,5))
species_count.plot(kind='pie',autopct='%1.1f%%',startangle=140)
plt.title("Specices Distribution")
#plt.xlabel("")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data['petal_length'],data['petal_width'],color = 'blue')
for i in range(len(data)):
    plt.text(data['petal_length'][i],data['petal_width'][i],data['species'][i]
            , fontsize = 8,ha='right')
    
plt.title("Scatter plot with annotations")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
plt.figure(figsize=(18, 14))
# First subplot - Line plot
plt.subplot(2, 2, 1)
plt.plot(data['petal_length'], color='green')
plt.title("Line Plot of Petal Length")
# Secnd subplot - Histogram
plt.subplot(2, 2, 2)
plt.hist(data['petal_width'], bins=15, color='orange')
plt.title("Histogram of Petal Width")
# Third subplot - Box plot
plt.subplot(2,2,3)
data.boxplot(column='sepal_length', by='species')
plt.title("Boxplot of Sepal Length by Species")
# Fourth subplot - Scatter plot

plt.subplot(2,2,4)
plt.scatter(data['sepal_length'], data['sepal_width'], color='purple')
plt.title("Scatter Plot of Sepal Length vs Width") 

plt.suptitle('')  # Suppresses the default title
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6)) 
# Adding diﬀerent styles and markers 
plt.plot(data['petal_length'], label='Petal Length', linestyle='-', marker='o', color='steelblue', linewidth=1.5) 
plt.plot(data['petal_width'], label='Petal Width', linestyle='--', marker='s', color='firebrick', linewidth=1.5) 
plt.plot(data['sepal_length'], label='Sepal Length', linestyle=':', marker='^', color='darkgreen', linewidth=1.5) 
plt.title("Enhanced Line Plot of Iris Dimensions") 
plt.xlabel("Index") 
plt.ylabel("Measurement (cm)") 
plt.legend() 
plt.grid(which='both', linestyle='--', linewidth=0.5) 
plt.show()
plt.savefig("plot_example.jpg",dpi=300)
```

```python
plt.figure(figsize=(10, 6)) 
# Use colormap to represent petal length 
scatter = plt.scatter( 
    data['sepal_length'], 
    data['sepal_width'], 
    c=data['petal_length'], 
    cmap='viridis', 
    s=data['petal_width'] * 50,  # Scale size based on petal width 
    alpha=0.7, 
    edgecolor='k' 
) 
plt.colorbar(scatter, label='Petal Length') 
plt.title("Styled Scatter Plot of Sepal Dimensions with Petal Characteristics") 
plt.xlabel("Sepal Length") 
plt.ylabel("Sepal Width") 
plt.grid(True)
```

```python
# Filter data by species 
setosa = data[data['species'] == 'setosa'] 
versicolor = data[data['species'] == 'versicolor'] 
virginica = data[data['species'] == 'virginica'] 
# Plot stacked histogram for sepal lengths 
plt.figure(figsize=(10, 6)) 
plt.hist( 
    [setosa['sepal_length'], versicolor['sepal_length'], virginica['sepal_length']], 
    bins=15, 
    stacked=True, 
    color=['#FF9999', '#66B2FF', '#99FF99'], 
    label=['Setosa', 'Versicolor', 'Virginica'] 
) 
plt.title("Stacked Histogram of Sepal Length by Species") 
plt.xlabel("Sepal Length (cm)") 
plt.ylabel("Frequency") 
plt.legend() 
plt.grid(axis='y', linestyle='--', linewidth=0.5) 
plt.show()
```
